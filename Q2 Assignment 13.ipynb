IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

READING THE DATASET

```python
data=pd.read_csv(r"C:\Users\Sujal Sahu\Downloads\forestfires.csv")
```

```python
data.drop(['month','day'],axis=1,inplace=True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
data['size_category']=label_encoder.fit_transform(data['size_category'])
data
```

SPLITTING THE DATASET INTO DEPENDENT AND INDEPENDENT VARIABLE

```python
x=data.iloc[:,0:28]
y=data.iloc[:,-1]
```

```python
x
```

```python
y
```

SPLITTING THE DATASET INTO TRAIN DATA AND TEST DATA

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

GRID SEARCH CV

```python
clf=SVC()
parameters={'kernel':['rbf'],
           'gamma':[100,50,5,0.5,0.1,0.01,0.0001,'auto'],
           'C':[50,15,10,6,5,0.1,0.001,0.0001]}
gsv=GridSearchCV(clf,param_grid=parameters,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_
```

```python
gsv.best_score_
```

MODEL FITTING USING SVC

```python
model=SVC(C=50,gamma=0.0001)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

MODEL ACCURACY

```python
acc=accuracy_score(y_test,y_pred)*100
print("Accuracy =",acc)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
